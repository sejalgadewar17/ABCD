#Z-score

```python
z = np.abs(stats.zscore(data['per']))
```

```python
print(z)
```

```python
threshold=0.18
```

```python
sample_outliers = np.where(z <threshold)
sample_outliers
```

#Inter Quantile Range(IQR)

```python
sorted_per= sorted(data['per'])
```

```python
sorted_per
```

```python
q1=np.percentile(sorted_per,25)
q3=np.percentile(sorted_per,75)
print(q1,q3)
```

```python
IQR=q3-q1
```

```python
lwr_bound=q1-(1.5*IQR)
upr_bound=q3+(1.5*IQR)
print(lwr_bound,upr_bound)
```

```python
p_outliers=[]
for i in sorted_per:
    if(i<lwr_bound or i>upr_bound):
        p_outliers.append(i)
print(p_outliers)
```

#handling outliers using Mean/Median imputation:

```python
col=['per']
data.boxplot(col)
```

```python
median=np.median(sorted_per)
median
```

```python
refined_df=data
refined_df['per'] = np.where(refined_df['per'] >upr_bound, median,refined_df['per'])
```

```python
refined_df
```

```python
refined_df['per'] = np.where(refined_df['per'] <lwr_bound, median,refined_df['per'])
```

```python
refined_df
```

```python
col = ['per']
refined_df.boxplot(col)
```

#To decrease the skewness and convert distribution into normal distribution

```python
new_df=data
for i in sample_outliers:
  new_df.drop(i,inplace=True)
new_df
```
